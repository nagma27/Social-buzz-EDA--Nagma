```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
content_df = r'C:\PROJECT\Accenture_internship\Content.csv'



# Define the file path

# Read the CSV file into a DataFrame
content_df = pd.read_csv( r'C:\PROJECT\Accenture_internship\Content.csv')



# Display the first few rows of the DataFrame to verify that it was loaded correctly
content_df.head()

     
```

```python
content_df.info()
     
```

```python
#Summary of the data
content_df.describe()
```

```python
content_df.drop(labels=['User ID','URL'],axis=1,inplace=True)
     
content_df.info()
```

```python
#Check values in category columns
content_df['Category'].value_counts()
     
```

```python
#strip quote mark and lowercase all values
content_df['Category'] = content_df['Category'].apply(lambda x:x.strip('"').lower())
     
```

```python
#checking if all categories are in lower case
content_df['Category'].value_counts()
     
```

```python
#rename content to content_type
content_df.rename(columns={'Type':'Content_Type'},inplace=True)
content_df['Content_Type'].value_counts()
```

```python
#import 2nd dataset
reactions_df = pd.read_csv(r'C:\PROJECT\Accenture_internship\Reactions.csv',index_col=0)
reactions_df.head()
```

```python
reactions_df.info()
```

```python
reactions_df.describe()
```

```python
# drop user ID column
reactions_df.drop(labels = ['User ID'],axis=1,inplace=True)
     
```

```python
# convert dates to datetime
reactions_df['Datetime'] = pd.to_datetime(reactions_df['Datetime'])
```

```python
#drop rows with missing values
reactions_df.dropna(inplace=True)
     
```

```python
#rename column
reactions_df.rename(columns={"Type":'Reaction_Type'},inplace=True)
     
```

```python
reactions_df.info()
```

```python
#import 3rd Dataset
reactiontypes_df = pd.read_csv(r'C:\PROJECT\Accenture_internship\ReactionTypes.csv')
reactiontypes_df.head()
     
```

```python
reactiontypes_df.info()
```

```python
reactiontypes_df.describe()
```

```python
#rename type column
reactiontypes_df.rename(columns={'Type':'Reaction_Type'},inplace=True)
reactiontypes_df.info()
```

```python
reactiontypes_df.value_counts()
```

# "Our data has been thoroughly cleansed in accordance with the client's requirements. This process involved removing any rows containing missing values, eliminating superfluous columns deemed irrelevant to the analysis, and ensuring consistency between data types and their respective column values."

# "Constructing a consolidated table to facilitate data modeling and analysis."(DATA MODELLING)

```python
#merge tables on content ID columns
top_performers = reactions_df.merge(content_df,on='Content ID')
top_performers
```

```python
#merge tables on reaction type column
top_performers = top_performers.merge(reactiontypes_df,on='Reaction_Type')
top_performers
```

# "Utilizing our dataset of 24,573 rows to identify the top five most popular categories."

# "Top Category Analysis: Exploring Categories Based on Popularity"

# "To determine category popularity, we sum the 'Score' column values by category."

```python
top_five = top_performers.groupby('Category').sum().sort_values(by='Score',ascending=False)
top_five.head(5)
```
